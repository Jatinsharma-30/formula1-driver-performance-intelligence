# Executive Summary

This project analyzed historical Formula 1 driver performance data to identify the characteristics of successful drivers and develop a predictive model for championship success.

The project combined SQL, Python, and Machine Learning to transform raw driver statistics into actionable insights.

The workflow included data validation, exploratory analysis, feature engineering, predictive modeling, and business recommendations.

# Key SQL Insights

### Driver Performance

- Elite drivers account for a significant share of total race victories.
- Championship-winning drivers consistently record higher podium finishes.
- Higher race participation alone does not guarantee championship success.

### Nationality Analysis

- A small number of countries have historically produced the majority of Formula 1 champions.
- Nationality trends highlight long-term dominance by established motorsport nations.

### Career Efficiency

- Drivers with higher win rates generally achieve greater championship success.
- Consistency is a stronger indicator of success than occasional victories.

# Exploratory Data Analysis Findings

- Race Wins and Podiums show a strong positive relationship.
- Pole Positions positively correlate with Race Wins.
- Career Points are highly concentrated among elite drivers.
- Performance varies across different Formula 1 eras.
- Championship drivers consistently outperform non-champions across key performance metrics.

# Most Influential Features

The model identified the following variables as the strongest predictors of championship success:

- Race Wins
- Podiums
- Points
- Pole Positions
- Years Active

These metrics consistently contributed to distinguishing champion and non-champion drivers.

# Business Recommendations

Based on the analysis:

- Focus on consistency rather than isolated race victories.
- Improve qualifying performance to increase race-winning opportunities.
- Monitor podium frequency as an early indicator of championship potential.
- Invest in long-term driver development to maximize career performance.
- Use performance analytics to support driver evaluation and strategic decision-making.
