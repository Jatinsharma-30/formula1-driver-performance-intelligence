# Formula 1 Driver Performance Intelligence

## Notebook 04: Feature Engineering

### Objective

This notebook creates new features from the existing Formula 1 driver dataset to improve machine learning performance and enhance analytical insights.

### Tasks Performed

- Load cleaned dataset
- Create derived features
- Encode categorical variables
- Scale numerical features
- Prepare final dataset for modeling

```python
import pandas as pd
import numpy as np

from sqlalchemy import create_engine

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
engine = create_engine("mysql+pymysql://root:root@localhost:3306/f1")

df = pd.read_sql("SELECT * FROM drivers", engine)

df.head()
```

```python
df["Championship_Experience"] = np.where(
    df["Champion"] == True,
    "Champion",
    "Non-Champion"
)

df.head()
```

```python
df["Wins_Per_Season"] = round(
    df["Race_Wins"] / df["Years_Active"],
    2
)

df.head()
```

```python
df["Podiums_Per_Season"] = round(
    df["Podiums"] / df["Years_Active"],
    2
)
```

```python
df["Points_Per_Season"] = round(
    df["Points"] / df["Years_Active"],
    2
)
```

```python
df["Win_Efficiency"] = round(
    df["Race_Wins"] / df["Race_Starts"],
    3
)
```

```python
df["Podium_Efficiency"] = round(
    df["Podiums"] / df["Race_Starts"],
    3
)
```

```python
df["Pole_Conversion"] = round(
    df["Race_Wins"] / df["Pole_Positions"],
    2
)

df["Pole_Conversion"] = (
    df["Pole_Conversion"]
    .replace([np.inf, -np.inf], 0)
    .fillna(0)
)
```

```python
encoder = LabelEncoder()

df["Nationality_Encoded"] = encoder.fit_transform(
    df["Nationality"]
)

df.head()
```

```python
scaler = StandardScaler()

scaled_columns = [
    "Race_Entries",
    "Race_Wins",
    "Podiums",
    "Points",
    "Years_Active"
]

df[scaled_columns] = scaler.fit_transform(
    df[scaled_columns]
)

df.head()
```

```python
df.info()
```

```python
df.to_csv(
    "../Results/engineered_dataset.csv",
    index=False
)

print("Feature engineered dataset saved successfully!")
```
